# Code to extract all files in Seperate three folders BG,CL,NM
BG : With Bag

CL : With Cloth

NM : Normal Walk

```python
# Importing necessary Libraries for data reading
import os
import shutil
```

### For BG

```python
dest_dir = './CASIA B_3/bg'
path ="./GaitDatasetB-silh"
#we shall store all the file names in this list
filelist = []
c=0
for root, dirs, files in os.walk(path):

        for file in files:
          if (file.find("bg")!=-1):
            #append the file name to the list
            filelist.append(os.path.join(root,file))
	
#print all the file names
for name in filelist:
    c=c+1
    print(name)
    shutil.copy(name,dest_dir)

print(c)
```

### For CL

```python
dest_dir = './CASIA B_3/cl'
path ="./GaitDatasetB-silh"
#we shall store all the file names in this list
filelist = []
c=0
for root, dirs, files in os.walk(path):

        for file in files:
          if (file.find("cl")!=-1):
            #append the file name to the list
            filelist.append(os.path.join(root,file))
	
#print all the file names
for name in filelist:
    c=c+1
    print(name)
    shutil.copy(name,dest_dir)
print(c)
```
